```python
import pandas as pd

dfs = [pd.read_csv(f"./multiTimeline{country}.csv") for country in ["PL", "US", "UK"]]
```

```python
for df in dfs:
    df.set_index("Miesiąc", inplace=True)
    

df = pd.concat(dfs, axis=1)
df.head()
```

```python
df.rename(columns={"vacation: (Polska)": "PL", "vacation: (Stany Zjednoczone)": "US", "vacation: (Wielka Brytania)": "UK"}, inplace=True)
```
